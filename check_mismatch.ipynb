```python
labels = sbm['pred_label']
# labels
```

```python
tags_sbj = test['subject_entity']
tags_obj = test['object_entity']
# print(tags_sbj)
# print(tags_obj)
```

```python
mismatched = []
for idx, (sbj, obj, label) in enumerate(zip(tags_sbj, tags_obj, labels)):
    if label != 'no_relation':
        sbj_tags = ['per', 'org']
        obj_tags = ['per', 'org', 'poh', 'dat', 'noh', 'loc']
        
        per_tags = ['per', 'org', 'poh', 'dat', 'noh', 'loc']
        org_tags = ['per', 'org', 'poh', 'dat', 'noh', 'loc']
        
        sbj_type_idx = sbj.index('type')
        obj_type_idx = obj.index('type')
        sbj_type = sbj[sbj_type_idx+8 : sbj_type_idx+11].lower()
        obj_type = obj[obj_type_idx+8 : obj_type_idx+11].lower()
        
        if (sbj_type not in sbj_tags) or (obj_type not in obj_tags):
            print(f"wrong indexing in index {idx}")
            print(sbj_type, obj_type)
        else:
            if sbj_type not in label:
                mismatched.append([idx, sbj_type, obj_type, label])
            else:
                if sbj_type == 'per':
                    if obj_type not in per_tags:
                        mismatched.append([idx, sbj_type, obj_type, label])
                elif sbj_type == 'org':
                    if obj_type not in org_tags:
                        mismatched.append([idx, sbj_type, obj_type, label])
mismatched
```
